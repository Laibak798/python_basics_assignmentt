```python
def check_subsequence(main_list, sub_list):
  it = iter(main_list)
  return all(element in it for element in sub_list)

main_list = [1,2,3,4,5,6]
sub_list = [2,5]
result = check_subsequence(main_list, sub_list)
print("Is sub_list a subsequence of main_list?",result)
```

```python
def fibonacci(n):
  fib_sequence = []
  a,b = 0,1
  while len(fib_sequence)<n:
    fib_sequence.append(a)
    a,b = b, a+b
  return fib_sequence

first_five_fib = fibonacci(5)
fib_set = set(first_five_fib)
print(fib_set)
```

```python
#50.Merge multiple sets into one.
```

```python
def merge_set(*sets):
  merged_set = set()
  for s in sets:
    merged_set = merged_set.union(s)
  return merged_set

set1 = {1,2,3}
set2 = {2,3,4}
set3 = {4,5,6}
merged_set = merge_set(set1,set2,set3)
print(merged_set)
```
